```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)

df.fillna(0, inplace=True)
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Month'] = df['Order_Date'].dt.month_name()
```

```python
total_revenue = df['Total_Sales'].sum()

total_orders = df['Order_ID'].nunique()

total_customers = df['Customer_ID'].nunique()

avg_order_value = total_revenue / total_orders

total_quantity = df['Quantity'].sum()

print("Total Revenue:", round(total_revenue,2))
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)
print("Average Order Value:", round(avg_order_value,2))
print("Total Quantity Sold:", total_quantity)
```

```python
monthly_sales = df.groupby('Month')['Total_Sales'].sum().reset_index()

fig = px.bar(
    monthly_sales,
    x='Month',
    y='Total_Sales',
    title='Monthly Sales Trend'
)

fig.show()
```

```python
category_sales = df.groupby('Category')['Total_Sales'].sum().reset_index()

fig = px.pie(
    category_sales,
    names='Category',
    values='Total_Sales',
    title='Category Contribution'
)

fig.show()
```

```python
product_sales = df.groupby(
    'Product'
)['Total_Sales'].sum().reset_index()

product_sales = product_sales.sort_values(
    by='Total_Sales',
    ascending=False
).head(10)

fig = px.bar(
    product_sales,
    x='Product',
    y='Total_Sales',
    title='Top 10 Products'
)

fig.show()
```

```python
city_sales = df.groupby(
    'City'
)['Total_Sales'].sum().reset_index()

city_sales = city_sales.sort_values(
    by='Total_Sales',
    ascending=False
).head(10)

fig = px.bar(
    city_sales,
    x='City',
    y='Total_Sales',
    title='Top Cities by Revenue'
)

fig.show()
```

```python
customer_sales = df.groupby(
    ['Customer_ID','Customer_Name']
)['Total_Sales'].sum().reset_index()

def segment(x):
    if x > 150000:
        return "High Value"
    elif x > 75000:
        return "Medium Value"
    else:
        return "Low Value"

customer_sales['Segment'] = customer_sales[
    'Total_Sales'
].apply(segment)

customer_sales.head()
```

```python
fig = px.pie(
    customer_sales,
    names='Segment',
    title='Customer Segmentation'
)

fig.show()
```

```python
gender_sales = df.groupby(
    'Gender'
)['Total_Sales'].sum().reset_index()

fig = px.bar(
    gender_sales,
    x='Gender',
    y='Total_Sales',
    title='Sales by Gender'
)

fig.show()
```

```python
fig = px.histogram(
    df,
    x='Age',
    title='Customer Age Distribution'
)

fig.show()
```
